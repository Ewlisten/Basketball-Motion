```python
# ---- Paths ----
MODEL_PATH   = "MODEL PATH"
VIDEO_PATH   = "VIDEO PATH"
OUTPUT_VIDEO = "OUTPUT PATH"

# ---- Detection ----
CONF_THRESH      = 0.05   # lower = more detections, more false positives
IMGSZ            = 640   # inference resolution (higher = better on small ball)

# ---- Ball tracking ----
TRAIL_LENGTH     = 35     # frames of fading dot trail behind ball
MIN_TRACK_LEN    = 3      # minimum detections to attempt a parabola fit
MAX_BALL_JUMP_PX = 160     # max px distance ball can move between frames without reset
BALL_MIN_Y_FRACTION = 0.01   # ignore balls detected in top 10% of frame (crowd/scoreboard)
BALL_MAX_Y_FRACTION = 0.09   # ignore balls below 90% (floor artifacts)

# ---- Shot detection ----
POST_APEX_FRAMES = 6      # how many frames after apex to accept rim overlap as a make
RIM_EXPAND_PX    = 20     # expand rim bbox by this many pixels for overlap check (forgiveness)
MIN_R_SQUARED    = 0.85   # min parabola fit to trust shot detection
RIM_SMOOTH_ALPHA = 0.35   #EMA weight for rim detection (lower = more smoothing)
MIN_ARC_HEIGHT_PX = 40    # Min vertical extent before ball is treated as a shot
# ---- Overlay style ----
OVERLAY_DURATION = 90     # frames to show MADE / MISSED banner
ARC_COLOR_MAKE   = (0,   220,  80)   # green
ARC_COLOR_MISS   = (0,    60, 220)   # red
ARC_COLOR_LIVE   = (255, 255, 255)   # white (arc color before outcome is known)
TRAIL_COLOR      = (255, 200,   0)   # amber ball trail
RIM_COLOR        = (0,   165, 255)   # orange rim box
```

```python
# Load trained weights (adjust paths based on your actual training output directories)
# Assuming 'train2' was the run for model_v5 (from Basketball-Motion,-5)
# And 'train3' is the run for model_v4 (from Basketball-Motion,-4) -- you might need to check your runs directory for the exact name
model_v5 = YOLO('/content/runs/detect/train3/weights/best.pt')
model_v4 = YOLO('/content/runs/detect/train5/weights/best.pt')

# What each model contributes
V5_CLASSES = ['player']           # trust V5 for players only
V4_CLASSES = ['ball', 'rim']      # trust V4 for ball and rim only

def predict_merged(source, conf_player=0.40, conf_ball=0.15):
    """
    Run both models and merge detections by class responsibility.
    Returns a list (one entry per frame) of detection dicts.
    """
    results_v5 = model_v5.predict(source=source, imgsz=640,  conf=conf_player, device='cpu', verbose=False)
    results_v4 = model_v4.predict(source=source, imgsz=960, conf=conf_ball,   device='cpu', verbose=False)

    merged_frames = []

    for r5, r4 in zip(results_v5, results_v4):
        frame_dets = []

        # Pull players from V5
        for box in (r5.boxes or []):
            name = model_v5.names[int(box.cls[0])]
            if name in V5_CLASSES:
                x1,y1,x2,y2 = map(int, box.xyxy[0].tolist())
                frame_dets.append({
                    "class":  name,
                    "conf":   float(box.conf[0]),
                    "xyxy":   [x1,y1,x2,y2],
                    "center": [int((x1+x2)/2), int((y1+y2)/2)],
                    "source": "v5"
                })

        # Pull ball + rim from V4
        for box in (r4.boxes or []):
            name = model_v4.names[int(box.cls[0])]
            if name in V4_CLASSES:
                x1,y1,x2,y2 = map(int, box.xyxy[0].tolist())
                frame_dets.append({
                    "class":  name,
                    "conf":   float(box.conf[0]),
                    "xyxy":   [x1,y1,x2,y2],
                    "center": [int((x1+x2)/2), int((y1+y2)/2)],
                    "source": "v4"
                })

        merged_frames.append(frame_dets)

    return merged_frames
```

```python
def get_video_info(path):
    """Return (fps, width, height, total_frames) for a video file."""
    cap = cv2.VideoCapture(path)
    fps    = cap.get(cv2.CAP_PROP_FPS) or 30.0
    w      = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    h      = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    total  = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.release()
    return fps, w, h, total
```

```python
def fit_parabola(frame_indices, ys, fps):
    """
    Fit a quadratic y(t) = a*t^2 + b*t + c to the ball's vertical pixel positions.
    Returns (poly_coefficients, poly1d_object, r_squared).
    """
    frames = np.array(frame_indices, dtype=float)
    times  = (frames - frames[0]) / fps
    ys_arr = np.array(ys, dtype=float)
    poly   = np.polyfit(times, ys_arr, 2)
    p       = np.poly1d(poly)
    ss_res  = np.sum((ys_arr - p(times)) ** 2)
    f = ys_arr # Original code snippet was missing ys_arr definition.
    ss_tot  = np.sum((ys_arr - np.mean(ys_arr)) ** 2)
    r_sq    = 1.0 - ss_res / ss_tot if ss_tot > 0 else 0.0
    return poly, p, times, r_sq
```

```python
def predict_arc_points(frame_indices, xs, ys, fps, extend_frames=20):
    """
    Return a list of (px_x, px_y) pixel points tracing the fitted parabola,
    extended `extend_frames` into the future for predictive display.
    """
    poly, p, times, _r_sq = fit_parabola(frame_indices, ys, fps)
    frames = np.array(frame_indices, dtype=float)
    t_end  = (frames[-1] - frames[0]) / fps + extend_frames / fps
    ts     = np.linspace(0, t_end, num=80)
    pts    = []
    for t in ts:
        py = int(p(t))
        # linear x interpolation; clamp beyond last observed frame
        t_clamped = min(t, times[-1])
        px = int(np.interp(t_clamped, times, np.array(xs, dtype=float)))
        pts.append((px, py))
    return pts, poly
```

```python
import math
import numpy as np

def is_made_shot(track, poly, fps, rim_bbox, times, xs):
    """
    Determines if a ball track represents a made shot.
    Handles jump shots (detected overlap) and dunks/layups (parabola extrapolation).

    Args:
        track  : list of (frame_idx, cx, cy, bbox, conf)
        poly   : np.polyfit coefficients [a, b, c] for y(t) = a*t^2 + b*t + c
        fps    : video frame rate
        rim_bbox : [x1, y1, x2, y2] of rim detection
        times  : relative time array matching track points (seconds)
        xs     : x positions matching track points

    Returns:
        (made: bool, method: str, frame: int or None)
    """
    a, b, c = poly
    x1, y1, x2, y2 = rim_bbox

    rim_cx = (x1 + x2) / 2
    rim_cy = (y1 + y2) / 2
    rim_w  = x2 - x1
    rim_h  = y2 - y1

    frames = np.array([t[0] for t in track])

    # ------------------------------------------------------------------ #
    # Guard: need some curvature to fit a meaningful arc                   #
    # Very small |a| means nearly linear motion (e.g. pass, not a shot)   #
    # ------------------------------------------------------------------ #
    if abs(a) < 1e-5:
        return False, "no_arc", None

    # ------------------------------------------------------------------ #
    # Find apex time                                                        #
    # ------------------------------------------------------------------ #
    t_apex = -b / (2 * a)

    # ------------------------------------------------------------------ #
    # METHOD 1 — Direct detection overlap (classic jump shot)              #
    # Ball is actually detected inside the rim bbox after the apex         #
    # ------------------------------------------------------------------ #
    after_idx = np.where(times >= t_apex)[0]

    for idx in after_idx:
        cx = track[idx][1]
        cy = track[idx][2]
        t  = times[idx]

        # Generous rim window: 1.2x width, 2x height (ball can enter from side)
        in_x = (cx >= x1 - rim_w * 0.1) and (cx <= x2 + rim_w * 0.1)
        in_y = (cy >= y1 - rim_h * 0.5) and (cy <= y2 + rim_h * 1.0)

        if in_x and in_y:
            vy = 2 * a * t + b  # dy/dt — positive = moving downward in image coords
            if vy > 0:
                return True, "detected_overlap", int(frames[idx])

    # ------------------------------------------------------------------ #
    # METHOD 2 — Parabola extrapolation (dunks, layups, occluded shots)   #
    # Solve for where the fitted arc crosses rim_cy, check if x is on hoop#
    # ------------------------------------------------------------------ #
    #
    # Solve: a*t^2 + b*t + c = rim_cy
    #     => a*t^2 + b*t + (c - rim_cy) = 0
    #
    discriminant = b**2 - 4 * a * (c - rim_cy)

    if discriminant >= 0:
        sqrt_disc = math.sqrt(discriminant)
        # Two solutions — we want the one where the ball is descending
        t_roots = [
            (-b + sqrt_disc) / (2 * a),
            (-b - sqrt_disc) / (2 * a),
        ]

        for t_rim in t_roots:
            # Must be after apex and within the track's time window
            t_max = times[-1]
            if t_rim < t_apex:
                continue
            # Allow extrapolation up to 0.5s beyond last detection
            # (covers occlusion during dunk / layup contact)
            if t_rim > t_max + 0.5:
                continue

            # Interpolate (or extrapolate) x at this time
            if t_rim <= t_max:
                x_at_rim = float(np.interp(t_rim, times, xs))
            else:
                # Linear extrapolation from last two x positions
                if len(xs) >= 2:
                    dt = times[-1] - times[-2]
                    if dt > 0:
                        vx = (xs[-1] - xs[-2]) / dt
                        x_at_rim = xs[-1] + vx * (t_rim - times[-1])
                    else:
                        x_at_rim = xs[-1]
                else:
                    x_at_rim = xs[-1]

            # Check if trajectory x lands over the rim
            # Use 1.4x rim width — ball center can be near edge and still go in
            tolerance = rim_w * 0.7  # half of 1.4x width
            if abs(x_at_rim - rim_cx) <= tolerance:
                # Confirm ball is moving downward at rim crossing
                vy_at_rim = 2 * a * t_rim + b
                if vy_at_rim > 0:
                    # Convert t_rim back to an approximate frame number
                    approx_frame = int(frames[0] + t_rim * fps)
                    return True, "extrapolated", approx_frame

    return False, "miss", None
```

```python
def draw_fading_trail(frame, trail_points, color=TRAIL_COLOR, max_radius=7):
    """Draw a fading dot trail. trail_points is a deque of (cx, cy) oldest→newest."""
    n = len(trail_points)
    for i, (cx, cy) in enumerate(trail_points):
        alpha  = (i + 1) / n
        radius = max(2, int(max_radius * alpha))
        bright = tuple(int(c * alpha) for c in color)
        cv2.circle(frame, (cx, cy), radius, bright, -1)
```

```python
def draw_arc(frame, arc_pts, color, thickness=2):
    """Draw a smooth polyline arc from a list of (x, y) points."""
    for i in range(len(arc_pts) - 1):
        p1, p2 = arc_pts[i], arc_pts[i + 1]
        # skip points outside frame bounds
        h, w = frame.shape[:2]
        if not (0 <= p1[0] < w and 0 <= p1[1] < h):
            continue
        if not (0 <= p2[0] < w and 0 <= p2[1] < h):
            continue
        cv2.line(frame, p1, p2, color, thickness, lineType=cv2.LINE_AA)
```

```python
def draw_make_miss_banner(frame, text, color, alpha=1.0):
    """Render a large centered MADE / MISSED banner with a semi-transparent background."""
    h, w = frame.shape[:2]
    font       = cv2.FONT_HERSHEY_DUPLEX
    font_scale = 2.8
    thickness  = 5
    (tw, th), baseline = cv2.getTextSize(text, font, font_scale, thickness)
    tx = (w - tw) // 2
    ty = int(h * 0.18)
    # dark pill background
    pad = 20
    overlay = frame.copy()
    cv2.rectangle(overlay,
                  (tx - pad, ty - th - pad),
                  (tx + tw + pad, ty + baseline + pad),
                  (20, 20, 20), -1)
    cv2.addWeighted(overlay, 0.55 * alpha, frame, 1 - 0.55 * alpha, 0, frame)
    # text
    cv2.putText(frame, text, (tx, ty), font, font_scale,
                tuple(int(c * alpha) for c in color), thickness, cv2.LINE_AA)
```

MAIN ANALYSIS PIPELINE

```python
import json
import math
def run_analysis(video_path="/content/Wemby_StpBK_Traject.mov", output_path="/content/Wemby_StpBK_analyzed.mov"):

    fps, W, H, total_frames = get_video_info(video_path)
    print(f"Video: {total_frames} frames @ {fps:.1f} fps  |  {W}x{H}")

    cap = cv2.VideoCapture(video_path)
    out = cv2.VideoWriter(output_path,
                          cv2.VideoWriter_fourcc(*"mp4v"),
                          fps, (W, H))

    # --- ByteTrack setup via supervision ---
    # ByteTrack expects a specific Detections object structure, so we'll collect
    # merged detections and then convert them into sv.Detections for players only.
    tracker        = sv.ByteTrack()
    trace_annotator = sv.TraceAnnotator(
        color        = sv.ColorPalette.DEFAULT,
        thickness    = 2,
        trace_length = TRAIL_LENGTH,
    )
    box_annotator = sv.BoxAnnotator(thickness=1)
    label_annotator = sv.LabelAnnotator(text_scale=0.4, text_thickness=1)

    # --- State ---
    ball_trail      = []          # list of (cx, cy) for fading trail
    ball_track_fi   = []          # frame indices of ball detections
    ball_track_xs   = []          # ball cx values
    ball_track_ys   = []          # ball cy values
    rim_bbox        = None        # most recent rim detection
    shot_arc_pts    = None        # current arc to draw
    arc_color       = ARC_COLOR_LIVE
    overlay_text    = None        # "MADE" or "MISSED"
    overlay_color   = (255, 255, 255)
    overlay_frames  = 0           # countdown
    shot_log        = []          # list of dicts saved to JSON

    frame_idx = 0

    # Pre-run merged prediction for all frames to simplify the loop
    print("Running merged predictions...")
    all_merged_detections = predict_merged(video_path, conf_player=0.40, conf_ball=0.15)
    print("Merged predictions complete.")

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # ── Get merged detections for this frame ───────────────────
        frame_merged_dets = all_merged_detections[frame_idx]

        # Prepare detections for ByteTrack (players only)
        player_detections_xyxy = []
        player_detections_confidence = []
        player_detections_class_id = []
        player_labels = []

        best_ball = None
        all_ball_dets = []
        current_rim_bbox = None

        for det in frame_merged_dets:
            if det["class"] == "player":
                player_detections_xyxy.append(det["xyxy"])
                player_detections_confidence.append(det["conf"])
                # Assuming 'player' is class_id 1 from V5_CLASSES, but it's better to map dynamically if model_v5.names is accessible
                player_detections_class_id.append(1) # Placeholder, adjust if necessary
                player_labels.append(f"player {det['conf']:.2f}")
            elif det["class"] == "ball":
                all_ball_dets.append((det["center"][0], det["center"][1], det["conf"]))
            elif det["class"] == "rim":
                # For the main analysis, we only need the rim bounding box
                current_rim_bbox = det["xyxy"]

        # Convert player detections to sv.Detections object
        detections = sv.Detections(
            xyxy=np.array(player_detections_xyxy),
            confidence=np.array(player_detections_confidence),
            class_id=np.array(player_detections_class_id)
        )

        # ── ByteTrack update (for players) ───────────────────────
        detections = tracker.update_with_detections(detections)

        # ── Process ball & rim from merged detections ────────────
        if all_ball_dets:
            if ball_trail:
                last_cx, last_cy = ball_trail[-1]
                candidates = [
                    (cx, cy, conf) for (cx, cy, conf) in all_ball_dets
                    if math.hypot(cx - last_cx, cy - last_cy) <= MAX_BALL_JUMP_PX
                ]
                if candidates:
                    best_ball = max(candidates, key=lambda x: x[2])[:2]
                else:
                    best_ball = max(all_ball_dets, key=lambda x: x[2])[:2]
                    ball_trail.clear()
                    ball_track_fi.clear()
                    ball_track_xs.clear()
                    ball_track_ys.clear()
            else:
                best_ball = max(all_ball_dets, key=lambda x: x[2])[:2]

        ball_det = best_ball

        if current_rim_bbox is not None:
            new_rim = list(map(float, current_rim_bbox))
            if rim_bbox is None:
                rim_bbox = new_rim
            else:
                rim_bbox = [
                    RIM_SMOOTH_ALPHA * new_rim[k] + (1 - RIM_SMOOTH_ALPHA) * rim_bbox[k]
                    for k in range(4)
                ]

        # ── Update ball trail and track ───────────────────────
        if ball_det:
            cx, cy = ball_det
            ball_trail.append((cx, cy))
            if len(ball_trail) > TRAIL_LENGTH:
                ball_trail.pop(0)

            ball_track_fi.append(frame_idx)
            ball_track_xs.append(cx)
            ball_track_ys.append(cy)

            # Keep track window: drop oldest points if track grows very long
            MAX_TRACK = 90
            if len(ball_track_fi) > MAX_TRACK:
                ball_track_fi = ball_track_fi[-MAX_TRACK:]
                ball_track_xs = ball_track_xs[-MAX_TRACK:]
                ball_track_ys = ball_track_ys[-MAX_TRACK:]

        # ── Fit parabola and check for shot when we have enough points ─────
        if len(ball_track_fi) >= MIN_TRACK_LEN:
            arc_height = max(ball_track_ys) - min(ball_track_ys)
            # Ensure enough points for polyfit (at least degree + 1, so 3 for degree 2)
            if len(ball_track_fi) >= 3:
                _, _, _, r_sq = fit_parabola(ball_track_fi, ball_track_ys, fps)
                if arc_height >= MIN_ARC_HEIGHT_PX and r_sq >= MIN_R_SQUARED:
                    arc_pts, poly = predict_arc_points(
                        ball_track_fi, ball_track_xs, ball_track_ys, fps
                    )
                    shot_arc_pts = arc_pts

                    # Check make/miss only if we have a rim and haven't just fired a banner
                    if rim_bbox is not None and overlay_frames == 0:
                        # Pass ball_track_fi as a list of tuples (frame_idx, cx, cy, bbox, conf) as expected by is_made_shot
                        # For simplicity, constructing a mock track with only required info for is_made_shot
                        mock_track = []
                        for i in range(len(ball_track_fi)):
                            mock_track.append((ball_track_fi[i], ball_track_xs[i], ball_track_ys[i], [0,0,0,0], 0.0)) # bbox and conf are placeholders

                        made, method, made_frame = is_made_shot(
                            mock_track, poly, fps, list(map(int, rim_bbox)), np.array([(fi - ball_track_fi[0]) / fps for fi in ball_track_fi]), ball_track_xs
                        )
                        if made_frame is not None:
                            label        = "MADE" if made else "MISSED"
                            arc_color    = ARC_COLOR_MAKE if made else ARC_COLOR_MISS
                            overlay_text  = label
                            overlay_color = arc_color
                            overlay_frames = OVERLAY_DURATION
                            shot_log.append({
                                "frame":      made_frame,
                                "result":     label,
                                "ball_pos":   [ball_track_xs[-1], ball_track_ys[-1]],
                                "rim_bbox":   list(map(int, rim_bbox)),
                            })
                            print(f"  Shot detected @ frame {made_frame}: {label}")
                            # Reset track so we don't re-trigger on same shot
                            ball_track_fi.clear()
                            ball_track_xs.clear()
                            ball_track_ys.clear()
                            shot_arc_pts = None


        # ── Draw supervision player boxes + traces ─────────────
        # The `labels` list must be created based on the `detections` object
        player_display_labels = [
            f"player {conf:.2f}"
            for conf in detections.confidence
        ]
        frame = trace_annotator.annotate(frame, detections)
        frame = box_annotator.annotate(frame, detections)
        frame = label_annotator.annotate(frame, detections, labels=player_display_labels)

        # ── Draw rim box ──────────────────────────
        if rim_bbox is not None:
            x1, y1, x2, y2 = map(int, rim_bbox)
            cv2.rectangle(frame, (x1, y1), (x2, y2), RIM_COLOR, 2)
            cv2.putText(frame, "RIM", (x1, y1 - 8),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.55, RIM_COLOR, 2)

        # ── Draw ball trail ───────────────────
        draw_fading_trail(frame, ball_trail)

        # ── Draw shot arc ───────────────
        if shot_arc_pts:
            draw_arc(frame, shot_arc_pts, arc_color)

        # ── Draw make/miss banner ───────────
        if overlay_frames > 0:
            # fade out in last 30 frames
            alpha = min(1.0, overlay_frames / 30.0)
            draw_make_miss_banner(frame, overlay_text, overlay_color, alpha)
            overlay_frames -= 1
            if overlay_frames == 0:
                shot_arc_pts = None
                arc_color    = ARC_COLOR_LIVE

        # ── Frame counter ───────────
        cv2.putText(frame, f"Frame {frame_idx}", (12, H - 14),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (180, 180, 180), 1)

        out.write(frame)
        frame_idx += 1

        if frame_idx % 100 == 0:
            print(f"  Processed {frame_idx}/{total_frames} frames...")

    cap.release()
    out.release()
    print(f"\nDone. Output saved to: {output_path}")

    # ── Save shot log to JSON ─────────────────
    log_path = str(Path(output_path).with_suffix(".json"))
    with open(log_path, "w") as f:
        json.dump({"video": Path(video_path).name, "fps": fps, "shots": shot_log}, f, indent=2)
    print(f"Shot log saved to: {log_path}")

    return shot_log
```
